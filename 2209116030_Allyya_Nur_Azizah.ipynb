##**Exploratory Data Analysis**

```python
df = pd.read_csv('laptops.csv')
price_counts = df['price(in Rs.)'].value_counts()
max_value = price_counts.max()

plt.figure(figsize=(8, 6))
sorted_data = df['price(in Rs.)'].sort_values().reset_index(drop=True)
plt.hist(sorted_data, bins=5, color='skyblue', edgecolor='black')

plt.title('Histogram Harga Produk')
plt.xlabel('Harga (Rupee)')
plt.ylabel('Frekuensi')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- plt.figure(figsize=(8, 6)): Ini adalah fungsi yang digunakan untuk membuat figur (gambar) baru dengan ukuran 8x6 inch menggunakan Matplotlib (plt).
- sorted_data = df['price(in Rs.)'].sort_values().reset_index(drop=True): Ini adalah operasi yang digunakan untuk mengurutkan nilai harga dalam kolom 'price(in Rs.)' dari DataFrame df secara ascending, kemudian disimpan dalam variabel sorted_data. Metode reset_index(drop=True) digunakan untuk mereset indeks DataFrame setelah diurutkan.
- plt.hist(sorted_data, bins=5, color='skyblue', edgecolor='black'): Ini adalah fungsi yang digunakan untuk membuat histogram dari data harga yang telah diurutkan (sorted_data). Parameter bins=5 menentukan jumlah bin yang digunakan dalam histogram. Parameter color='skyblue' menentukan warna isi histogram, dan edgecolor='black' menentukan warna garis tepi histogram.
- plt.title('Histogram Harga Produk'): Ini adalah fungsi yang digunakan untuk menambahkan judul pada plot menggunakan Matplotlib (plt). Judul yang ditambahkan adalah 'Histogram Harga Produk'.

- labels={'display(in inch)': 'Display', 'price(in Rs.)': 'Price', 'rating': 'Rating'}: Penamaan label sumbu dan warna yang digunakan pada plot. Parameter ini digunakan untuk memberi label yang lebih deskriptif pada sumbu-sumbu dan warna pada plot.

Ketika tidak ada nilai yang diduplikasi dalam dataset maka:
- Data memiliki keunikan di setiap barisnya.
- Tidak perlu dilakukan tindakan lebih lanjut terkait penghapusan atau pengelolaan nilai duplikat.
- Data ini mungkin telah melalui proses penghapusan nilai duplikat sebelumnya atau telah diambil dari sumber yang telah membersihkan data duplikat.
